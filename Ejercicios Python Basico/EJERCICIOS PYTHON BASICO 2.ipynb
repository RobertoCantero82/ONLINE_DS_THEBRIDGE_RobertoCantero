## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
import math
grados = float(input("Dime los grados y te los paso a radianes: "))
radianes = round(grados * (math.pi / 180), 2)
print("Los", grados,"convertidos serían", radianes,"radianes")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
base = int(input("¿Cuál es la base? (en centímetros y sin decimales): "))
altura = int(input("¿Cuál es la altura? (en centímetros y sin decimales): "))
area = base * altura
print("El paralelogramo de", base,"x", altura,"tiene un área de", area)
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
base = float(input("¿Cuál es la base? (en centímetros y con decimales): ").replace(",", "."))
altura = float(input("¿Cuál es la altura? (en centímetros y con decimales): ").replace(",", "."))
area = round(base * altura, 2)
print("El paralelogramo de", base,"x", altura,"tiene un área de", area)
```
