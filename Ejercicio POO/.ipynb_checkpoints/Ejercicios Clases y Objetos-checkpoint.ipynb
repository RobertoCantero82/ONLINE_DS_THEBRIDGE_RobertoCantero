## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
# Implementar la clase "Tienda" y los métodos

class Tienda():
    """Creamos la clase Tienda, que será, por defecto, de tipo electrodomésticos y abierta"""
    Tipo = "Electrodomésticos"
    Abierta = True
    def __init__(self, nombre, direccion, num_empleados, ventas_3meses):
        """Esta clase tendrá objetos con nombre, dirección, número de empleados y una lista con las ventas de los últimos 3 meses"""
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_3meses = ventas_3meses

    def ventas_totales(self):
        """Sumo las ventas de la lista ventas_3meses"""
        return sum(self.ventas_3meses)
    
    def media_ventas_empleado(self):
        """Divido ventas_totales por el número de empleados"""
        total_3meses = self.ventas_totales()
        return total_3meses / self.num_empleados
    
    def nombre_direccion(self):
        """Devuelvo el nombre y dirección de una tienda"""
        return f"{self.nombre} - {self.direccion}"
    
    def ventas_ultimo_mes(self):
        """Obtengo las ventas del último mes. Es decir, el último valor de la lista ventas_3meses"""
        return self.ventas_3meses[-1]
    
   # def proyeccion_ventas(self, x):
        # if x < 0:
           # return "Tienes que escribir una cifra positiva."    
       # elif x < 1000:
           # nuevalista = []
           # for i in self.ventas_3meses:
               # nuevalista.append(i * 1.2)
           # self.ventas_3meses = nuevalista
       # else:
           # nuevalista = []
           # for i in self.ventas_3meses:
               # nuevalista.append(i * 1.5)
           # self.ventas_3meses = nuevalista
       # return sum(self.ventas_3meses)

    # He tenido que rehacer el método, ya que al realizar el cálculo para la tienda1, tanto para un gasto de marketing de
    # más de 1000 euros, como de menos de 1000 euros, los valores se sumaban, modificaban la lista original de ventas de los últimos 3 meses
    # y esto hacía que el método de ventas totales, el de ventas por empleado y el de ventas del último mes se actualizasen también.
    # Por lo tanto, no reescribo las ventas y simplemente dejo creada una nueva lista.

    def proyeccion_ventas(self, x):
        """Da como resultado el total de ventas si se hiciesen acciones de marketing. Es la suma de una nueva lista"""
        if x < 0:
            return "Tienes que escribir una cifra positiva."    
        elif x < 1000:
            nuevalista = []
            for i in self.ventas_3meses:
                nuevalista.append(i * 1.2)
            return sum(nuevalista)
        else:
            nuevalista = []
            for i in self.ventas_3meses:
                nuevalista.append(i * 1.5)
            return sum(nuevalista)
        
        
```

```python
# Crear tres tiendas con datos inventados

tienda1 = Tienda(nombre = "ElectroUno", direccion = "Avenida de las avenidas, 1", num_empleados = 8, ventas_3meses = [35000, 32000, 40000])
tienda2 = Tienda(nombre = "ElectroDos", direccion = "Calle callejuela, 7", num_empleados = 12, ventas_3meses = [20000, 25000, 21000])
tienda3 = Tienda(nombre = "ElectroTres", direccion = "Avenida de las calles, 187 Bis", num_empleados = 32, ventas_3meses = [120000, 100000, 150000])
```

```python
# Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)

print("Esta es una tienda de",Tienda.Tipo,"y es",Tienda.Abierta,"que está abierta. Su nombre es",tienda1.nombre,", está en la dirección:", tienda1.direccion,", tiene ",tienda1.num_empleados,"empleados y unas ventas totales en los últimos tres meses de",tienda1.ventas_3meses," euros.")
print("*************************")

print("Las ventas de los últimos tres meses de la tienda 1 son de",tienda1.ventas_totales(),"euros.")
print("*************************")

print("La media de ventas de cada empleado de la tienda 1 es de",tienda1.media_ventas_empleado(),"euros.")
print("*************************")

print("El nombre y la dirección de la tienda 1 son: ",tienda1.nombre_direccion())
print("*************************")

print("Las ventas del último mes han sido de",tienda1.ventas_ultimo_mes(),"euros.")
print("*************************")

print("Las ventas proyectadas si se invierten 1500 euros en la tienda 1 serían de:",tienda1.proyeccion_ventas(1500))
print("*************************")

print("Las ventas proyectadas si se invierten 500 euros en la tienda 1 serían de:",tienda1.proyeccion_ventas(500))
print("*************************")
```

```python
# Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`

tiendas = [tienda1, tienda2, tienda3]
total = 0

for tienda in tiendas:
    total = total + tienda.ventas_ultimo_mes()

print("Las ventas totales de las", len(tiendas),"tiendas han sido de",total,"euros.")
    
    
```

```python
# Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"

tiendas = [tienda1, tienda2, tienda3]
palabra = "Avenida"

for tienda in tiendas:
    if palabra in tienda.direccion:
        print(tienda.nombre)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
# Implementa la clase Perro

class Perro():
    """Creamos la clase Perro que siempre tendrá 4 patas, 2 orejas y 2 ojos"""
    patas = 4
    orejas = 2
    ojos = 2

    def __init__(self, raza, dueño, pelo = "Marrón", velocidad = 0):
        """Creamos atributos como la raza, el dueño, el pelo, que por defecto será Marrón, el pelo y la velocidad, que por defecto será 0."""
        self.raza = raza
        self.dueño = dueño
        self.pelo = pelo
        self.velocidad = velocidad
        
    def andar(self, aumento_velocidad):
        """Para que el perro ande, sumo la velocidad marcada inicialmente al aumento asignado"""
        self.velocidad = self.velocidad + aumento_velocidad
        return self.velocidad

    def parar(self):
        """Para que el perro se pare, simplemente pongo el valor del atributo velocidad en 0"""
        self.velocidad = 0

    def ladrar(self, string):
        """Para que el perro ladre, utilizo el string GUAU! junto con un string de la elección del usuario"""
        return ("GUAU!" + string)
        
        
```

```python
# Crea un objeto de tipo Perro, sin dueño

schnauzer = Perro(raza = "Schnauzer", dueño = None, pelo = "sal y pimienta")
```

```python
# Comprueba que están bien todos sus atributos

print("Este perro tiene",Perro.patas,"patas y",Perro.orejas,"orejas. Por supuesto,",Perro.ojos,"ojos. Va a una velocidad de",schnauzer.velocidad,". Es de raza",schnauzer.raza,", su dueño es",schnauzer.dueño,"y su pelo es de color",schnauzer.pelo,".")
```

```python
# Prueba que ande, y comprueba su velocidad

schnauzer.andar(10)
print("Ahora, el perro anda a una velocidad de",schnauzer.velocidad)
```

```python
# Páralo

schnauzer.parar()
print("Ahora el schnauzer se ha parado, con lo que su velocidad es:",schnauzer.velocidad)
```
