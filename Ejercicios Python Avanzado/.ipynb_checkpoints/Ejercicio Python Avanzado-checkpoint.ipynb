## Ejercicio 6
Multiplica todos los valores del diccionario
> ```ejer_6 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

`````python
ejer_6 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}

resultado = 1 # Iniciamos la multiplicación con 1 (si fuese 0 no tendría sentido)

for i in ejer_6.values(): # Iteramos cada valor del diccionario
    resultado = resultado * i # En cada iteración, multiplicamos el resultado por el valor correspondiente. La primera vez
                              # será 1 * 11, que da 11. Luego 11 * 22, que da 242. Luego 242 * 33...etc  
print("El resultado de la multiplicación es:",resultado)
    
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
# Se crea la función de validación del DNI

def val_DNI(dni):
    # Para que el DNI sea válido se tiene que cumplir que tenga entre 7 y 8 dígitos numéricos
    # y que la letra se corresponda. Por tanto, hay que diferenciar entre los dígitos y la letra
    
    # Almacenamos la letra como variable (upper para que si se introduce en minúscula se ponga en mayúscula)
    letra = dni[-1].upper()
    
    # Almacenamos los dígitos como variable
    digitos = dni[:-1]
    
    # Lo primero que comprobamos es si el dni, salvo la ultima letras son digitos. Si no son digitos, no seguimos
    if not digitos.isdigit():
        return False
        
    # Comprobamos si son entre 7 y 8 digitos. Si no lo son, paramos.
    if not (7 <= len(digitos) >= 8):
        return False
        
    # Comprobamos que la letra es correcta    
    palabra ='TRWAGMYFPDXBNJZSQVHLCKE'
    numero = int(digitos) # Ya que digitos era un string y queremos un int para operar con el
    letra_correcta = palabra[numero % 23]
    
    # Si se han cumplido todos los pasos, ya podremos decir si el DNI es válido
    return letra == letra_correcta
`````
