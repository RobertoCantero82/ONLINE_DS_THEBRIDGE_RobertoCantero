### 18. ¿Cuál es la edad menos frecuente?

```python
'''
Aquí lo que estoy haciendo es utilizar el método value_counts() para obtener la frecuencia de cada valor único
en la columna "age". Después uso el método idxmin() para obtener el valor menos frecuente, es decir, el valor que menos aparece.

Sin embargo, aquí me encontraré con un problema que desarrollo debajo de esta celda.
'''
print(f"La edad menos frecuente es de: {users['age'].value_counts().idxmin()} años.")
```

En este caso, a través de **idxmin()** obtengo una solución única y dado que el dataset tiene casi 1000 entradas, parece raro que solamente una edad sea la menos frecuente. Por ello, aquí debajo comprobaré las edades y sus frecuencias. En caso de que haya más de un resultado con frecuencia igual a 1, **vuelto a realizar el ejercicio** para obtener no la edad menos frecuente, sino las **edades menos frecuentes**.

```python
'''
Obtengo el conteo de los valores menos frecuentes, ya que hay más de uno, como se puede ver aquí debajo. Según el método
value_counts(), las edades 7, 66, 11, 10 y 73 solo aparecen una vez en la columna "age".
'''
conteo = users["age"].value_counts()
print("Dado que hay más de una edad menos frecuente, primero obtengo el conteo de edades y frecuencias, que es:")
print()
conteo
print(conteo)
```

```python
'''
Creo una variable que incluya una máscara. Dentro de conteo, me quedo con las edades menos frecuentes, esto lo hago 
a través de conteo == conteo.min(). Y en la misma variable, con el atributo index, obtengo los índices de esas frecuencias,
que, en realidad, son las edades que estoy buscando.

Para que quede más bonito lo imprimo en forma de lista.

Por último, compruebo que las edades menos frecuentes son las mismas que los índices cuyos valores son iguales a 1.
Y esto lo hago también a través de una máscara: conteo == 1].index).
'''
edades_menos_frecuentes = conteo[conteo == conteo.min()].index
print()
print(f"Y el conjunto de edades menos frecuentes es: {list(edades_menos_frecuentes)}")
print()
print(f"Para comprobarlo, solo tengo que buscar aquellas edades que tienen un conteo igual a 1.")
print(f"Estas son: {list(conteo[conteo == 1].index)}. Y coincide con el resultado anterior.")
```
