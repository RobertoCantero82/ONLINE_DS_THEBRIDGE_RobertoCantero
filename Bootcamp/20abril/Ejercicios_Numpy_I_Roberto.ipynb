## Ejercicio 5
En un nuevo array, sustituye todos los impares de `my_array` por -1

````python
nuevo_array = []
for i in my_array:
    if i in impares:
        nuevo_array.append(-1)
    else:
        nuevo_array.append(i)

print(nuevo_array)

        
````

````python
# como me salía ese resultado raro arriba, he mirado que esto se debe a que mezclo listas y arrays
# la solución que he encontrado es convertir a int los elementos del array, con lo que así puedo
# ver los elementos como han quedado en el array de la sustitución

nuevo_array = []
for i in my_array:
    if i in impares:
        nuevo_array.append(-1)
    else:
        nuevo_array.append(int(i))

print(nuevo_array)
````

## Ejercicio 6
En un nuevo array, sustituye todos los impares de `my_array` por -1. Esta vez usa `where`

````python
nuevo_array = np.where(my_array % 2 != 0, -1, my_array) 

'''
He utilizado el método where, que no necesita de un bucle para hacer la sustitución.
Se invoca con np.where y consiste en tres parámetros:
    - El primero es la condición, en este caso que el elemento en cuestión sea impar, es decir, que su resto sea distinto de cero
    - El segundo es el valor que queremos sustituir (-1)
    - El tercero es donde queremos hacer la operación (en el array "nuevo_array")
'''

print(nuevo_array)
````

## Ejercicio 8
Concatena los siguientes arrays verticalmente

```Python
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)
```

````python
a = np.arange(10).reshape(2,-1) # con el reshape 2, -1, le decimos a NumPy que cree dos filas como quiera de un array inicial de 0 a 9
b = np.repeat(1, 10).reshape(2,-1) # dos filas como quiera, de diez unos

print(a) # imprimo el array a
print("-" *60)
print(b) # imprimo el array b
print("-" *60)

# Para concatenar verticalmente uso np.concatenate((arrays en formato tupla), eje   )

print(np.concatenate((a,b))) # concateno verticalmente con el método concatenate
print("-" *60)

# También se puede utilizar np.vstack((arrays en tupla))

print(np.vstack((a,b))) # concateno verticalmente con el método vstack
````

## Ejercicio 9
Concatena los siguientes arrays horizontamente

```Python
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)
```

````python
a = np.arange(10).reshape(2,-1) # array del 0 al 9 en dos dimensiones
b = np.repeat(1, 10).reshape(2,-1) # array de diez unos en dos dimensiones
print(a) # imprimo el array a
print("-" *60)
print(b) # imprimo el array b
print("-" *60)
print(np.concatenate((a,b), axis=1)) # concateno horizontalmente con el método concatenate
print("-" *60)
print(np.hstack((a,b))) # concateno horizontalmente con el método hstack
````

## Ejercicio 20
Dado el siguiente array

```Python
x = np.random.randint(10, 20, size = (5,2,6))
```

1. ¿Cuántos elementos tiene el array?
2. ¿Cuántas dimensiones?
3. ¿Cuánta es la longitud cada dimensión?

````python
x = np.random.randint(10, 20, size = (5,2,6))
print(x)
print("-" * 60)
print(f"El array tiene {x.size} elementos.") # x.size calcula los elementos del array
print(f"El array tiene {x.ndim} dimensiones.") # x.ndim calcula el nº de dimensiones
print(f"Cada dimensión tiene una longitud de {x.shape}.") # x.shape calcula la longitud de las dimensiones
````
