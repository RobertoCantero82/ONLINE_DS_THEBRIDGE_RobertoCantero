## 21. Pintamos el swarmplot con el melted y entendemos la información

### De un vistazo se ve cómo se distribuye cada estadística para los tres tipos: Agua (azul) tiende a tener más HP, Fuego (naranja) destaca en Ataque y Sp. Atk, y Hierba (verde) es más equilibrado, pero más bajo en casi todo.

## 22. Basándote en el gráfico anterior, define el tamaño de la figura en (10,6) (usa plt). En el swarmplot añade el argumento dodge a True y usa la anterior paleta de colores Pokemon. Dodge en inglés es "esquivar", ¿qué diferencia ves?

## Usando plt, fija el límite del eje Y entre 0 y 200 y fija la leyenda con bbox_to_anchor en upper left. Pista: upper left es el código 2 y se usa el parámetro loc.


```python
plt.figure(figsize=(10, 6))  # tamaño de la figura
sns.swarmplot(x="New Column", y="value", data=pokemon_melt, hue="Type 1", palette=colores, dodge=True)  # dodge separa los puntos por tipo y uso los colores anteriores
plt.ylim(0, 200)  # límite del eje Y
plt.legend(bbox_to_anchor=(1, 1), loc=2);  # leyenda 
```
