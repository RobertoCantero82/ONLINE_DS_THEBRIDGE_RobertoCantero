2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0,5) # para el eje x generamos un array de 0 a 5

y1 = x * 3 # el primer caso, el eje y tendrá los elementos multiplicados por 3

y2 = (-3 * x) -5 # aquí el eje y 

plt.subplot(1, 2, 1) # habrá una fila y dos columnas y este será el primer gráfico
plt.plot(x, y1) # creo el gráfico con x e y1

plt.subplot(1, 2, 2) # habrá una fila y dos columnas y este será el segundo gráfico
plt.plot(x,y2) # creo el gráfico con x e y2
```

3. Fija el eje X entre 0 y 5 y el eje Y entre -15 y 15

```python
x = np.linspace(0,5)

y1 = x * 3
y2 = -3 * x - 5

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.xlim(0, 5) # añado el límite de x (entre 0 y 5)
plt.ylim(-15, 15) # añado el límite de y (entre -15 y 15)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.xlim(0, 5) # añado el límite de x (entre 0 y 5)
plt.ylim(-15, 15) # añado el límite de y (entre -15 y 15)
```

4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
x = np.linspace(0,5)

y1 = x * 3
y2 = -3 * x - 5

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.xlim(0, 5)
plt.ylim(-15, 15)
plt.xlabel("eje X")      # Etiqueta eje X
plt.ylabel("eje Y")      # Etiqueta eje Y 
plt.title("recta sube")  # Título

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.xlim(0, 5)
plt.ylim(-15, 15)
plt.xlabel("eje X")      # Etiqueta eje X
plt.title("recta baja")  # Título
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0, 5)

y1 = 3 * x
y2 = -3 * x - 5

fig, (ax_izq, ax_dch) = plt.subplots(1, 2, figsize=(10, 4)) # creo una figura con dos subplots en horizontal

ax_izq.plot(x, y1)
ax_izq.set_xlim(0, 5)
ax_izq.set_ylim(-15, 15)

ax_dch.plot(x, y2)
ax_dch.set_xlim(0, 5)
ax_dch.set_ylim(-15, 15)
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html


Se puede hacer a través de un dataframe con una columna de alturas aleatorias:

```python
alturas = np.random.uniform(150, 190, 300) # genero alturas aleatorias

df_alturas = pd.DataFrame(alturas, columns=['Altura']) # creo el dataframe

df_alturas # muestro el dataframe
```

```python
df_alturas.hist(bins=20); # muestro el histograma con agrupaciones de 20
```

También se podría hacer directamente con el array aleatorio de alturas:

```python
alturas = np.random.uniform(150, 190, 300) # genero alturas aleatorias

plt.hist(alturas, bins=30); # muestro el histograma a partir del array de alturas, con bins de 30 elementos
```
