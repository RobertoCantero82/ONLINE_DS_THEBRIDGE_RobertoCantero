### Step 3. Name of the max valued product

```python
'''
Entiendo que me está pidiendo devolver el producto con el precio más alto. Aquí el problema es que la columna item_price
no tiene int, sino strings, ya que incluye el símbolo del dólar. Lo primero que habría que hacer es quitar el símbolo para poder
realizar operaciones con los precios. 
Leyendo documentación, si utilizase replace("$", ""), buscaría en las celdas exactamente el símbolo. Y lo que quiero no es eso, 
sino entrar al string y cambiar solo ese símbolo. Para ello se utiliza .str antes del replace.
'''

chipo["item_price"] = chipo["item_price"].str.replace("$", "")
print("Primero quito el símbolo $ en todos los valores de la columna item price:")
print()
chipo["item_price"]
```

```python
# Ahora convierto el formato de las celdas, de string a float

chipo["item_price"] = chipo["item_price"].astype(float)
print(f"El tipo de la columna item price ahora es: {chipo["item_price"].dtype}")
```

```python
# Ahora ya puedo operar con los valores de la columna y encontrar el indice con el precio máximo

print(f"El índice del producto con el precio más alto es: {chipo["item_price"].idxmax()}.")
```

```python
# Por último, busco el nombre del producto. Para ello uso el índice y busco en la columna item_name
# Como busco por nombre, tengo que usar loc (fila = indice, columna = item_name)

productomascaro = chipo.loc[chipo["item_price"].idxmax(), "item_name"]
print(f"El producto con el precio máximo es: {productomascaro}.")
```

### Step 4. How many products cost more than $10.00?

```python
# Tengo que buscar en la columna item_price las filas donde el valor sea mayor de 10

print(f"Hay {chipo[chipo["item_price"] > 10].shape[0]} productos que valen más de 10 dólares.")

# La máscara sería chipo["item_price"] > 10. Así obtengo los valores superiores a 10.
# Después simplemente utilizo el atributo shape, que me da número de filas y columnas y extraigo las filas que han
# dado positivo con la máscara.
```

### Step 4.1: Y cuántos pedidos se han hecho con un producto de más de 10$? Es lo mismo?

```python
# No sería lo mismo, ya que en un mismo pedido puede haber varios productos de más de 10 dólares. 
# Lo lógico sería que hubiese menos pedidos (mínimo los mismos que productos)
# Para poder contar los pedidos, lo que se busca es ver los valores únicos de la fila order_id con la máscara anterior.

prodmas10 = chipo[chipo["item_price"] > 10] # creo una variable para la máscara de productos de más de 10 dólares

print(f"Se han realizado {prodmas10["order_id"].nunique()} pedidos con productos de más de 10 dólares.")

# A través del método nunique() cuento los valores únicos en la columna order_id
```

### Step 4.2: Y cuántos pedidos se han hecho de más de 20$? Es lo mismo?

```python
# Habría que crear una nueva máscara para obtener los valores en item_price de más de 20
# Después, con unique(), obtener los valores únicos en la columna order_id

prodmas20 = chipo[chipo["item_price"] > 20]
print(f"Se han realizado {prodmas20["order_id"].nunique()} pedidos con productos de más de 20 dólares.")
```

### Step 4.3: Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? Es lo mismo?

```python
'''
Aquí habría que usar una máscara similar a la de los pasos 4 y 4.1, pero tenemos que irnos a la columna de quantity.
Además, antes hablábamos de precio unitario. Ahora queremos saber si en un pedido se pago más de 10 por un producto, es decir,
si la multiplicación de la columna quantity y la columna item_price es mayor a 10.
Por tanto hay que crear una nueva columna llamada total_price, que sera item_price * quantity
'''

chipo["total_item_price"] = chipo["item_price"] * chipo["quantity"]
print("Obtenemos el dataframe con la nueva columna \"total_item_price\":")
print()
chipo.head(10)
```

```python
pedidos10 = chipo[chipo["total_item_price"] > 10]["order_id"].nunique()
pedidos10
print(f"En {pedidos10} pedidos se han pagado más de 10 dólares por un mismo producto")
```

### Step 5. What is the price of each item and name it unit_price. Get only item_name and unit_price

```python
'''
Aquí entiendo que tengo que crear un nuevo dataframe a partir del anterior. Solo tiene que tener la columna de item_name y de unit_price.
Esa última columna no es nueva, sino simplemente la anterior item_price renombrada. Además, quiero que me muestre precio de cada producto, sin duplicados.
Habría que tener en cuenta que un mismo producto puede tener diferentes ingredientes. Con lo que habría que buscar duplicados y descartar los que
tengan el mismo precio, dejando los que tengan precios distintos. Por ejemplo, el Chicken Bowl puede costar 16,98 o 10,98 dependiendo de sus ingredientes.
'''

df_precios_unicos = chipo[["item_name", "item_price"]] # primero selecciono las columnas que quiero visualizar
df_precios_unicos
```

```python
df_precios_unicos = df_precios_unicos.rename(columns={"item_price":"unit_price"}) # renombro por columna, pasando nombre anterior y nuevo
print("Así quedaría el dataframe con el nuevo nombre de columna:")
print()
df_precios_unicos
```

```python
df_precios_unicos = df_precios_unicos.drop_duplicates() # utilizo el método drop_duplicates para eliminar los duplicados
print("Ahora elimino los duplicados, para que me quede lo que se pedía:")
print()
df_precios_unicos
```

### Step 6. Sort by the name of the item

```python
df_precios_unicos = df_precios_unicos.sort_values("item_name") # ordeno a través de la columna item_name
print("El dataframe ordenado por el nombre del producto sería:")
print()
df_precios_unicos
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
df_prod_mas_caro = chipo.sort_values("item_price", ascending=False) # primero ordeno el dataframe original por item_price en orden descendente
print("Este es el dataframe original ordenado de manera descendente por precio unitario:")
print()
df_prod_mas_caro
```

```python
print(f"{productomascaro} fue pedido {df_prod_mas_caro.iloc[0]['quantity']} veces.")
```

```python
'''
Anteriormente habíamos calculado productomascaro = chipo.loc[chipo["item_price"].idxmax(), "item_name"].
Ahora habría que hacer lo mismo, pero en lugar de obtener el producto, queremos obtener la cantidad pedida
'''

cantidad_mas_caro = chipo.loc[chipo["item_price"].idxmax(), "quantity"] # En lugar de item_name, pedimos la cantidad
print(f"{productomascaro} se pidió {cantidad_mas_caro} veces.")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
mascaraveggie = chipo[chipo["item_name"] == "Veggie Salad Bowl"] # creo la máscara que filtra en la columna item_name por el nombre dado
vecesveggie = mascaraveggie.shape[0] # busco la posición 0 del método shape para obtener el número de filas en las que aparece el elemento filtrado
print(f"El item Veggie Salad Bowl se pidió {vecesveggie} veces.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
mascarasoda = chipo[(chipo["item_name"] == "Canned Soda") & (chipo["quantity"] > 1)] # creo la máscara que cumpla que la columna item_name sea Canned Soda
                                                                                   # y que la cantidad sea más de 1
vecessoda = mascarasoda.shape[0] # obtengo las filas que coinciden con el filtrado realizado
print(f"Las veces que alguien ha pedido más de una Canned Soda son {vecessoda}.")
```
