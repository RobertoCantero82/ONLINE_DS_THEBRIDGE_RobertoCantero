## Ejercicio 1
Dada la siguiente lista:
> ```ejer_1 = [1,2,3,4,5]```

Inviértela para que quede de la siguiente manera
> ```ejer_1 = [5,4,3,2,1]```

````python
ejer_1 = [1,2,3,4,5]
ejer_1.reverse()
print(f"La lista ejer_1 es {ejer_1}")
# Otra manera con slicing (hecho en corrección de clase)
ej = [1,2,3,4,5]
ej = ej[::-1]
print(f"La lista ej es {ej}")
# Otra manera con sort (hecho en corrección de clase)
ej2 = [1,2,3,4,5]
ej2.sort(reverse=True)
print(f"La lista ej2 es {ej2}")
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
print(f"['{ejer_3_1[0]} {ejer_3_2[0]}', '{ejer_3_1[0]} {ejer_3_2[1]}', '{ejer_3_1[1]} {ejer_3_2[0]}', '{ejer_3_1[1]} {ejer_3_2[1]}']")
print("_" * 50)
print(" ")
# También se puede hacer con bucle (hecho en corrección de clase)
lista_final = []
for i in ejer_3_1:
    for j in ejer_3_2:
        lista_final.append(i + " " + j)
print(lista_final)
````

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
solucion = []
for i in ejer_5:
    if i != 3:
        solucion.append(i)
print(f"La lista final añadiendo a una lista vacía es {solucion}")

# También se puede iterar sobre una copia de la lista (hecho en clase en la correción)
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
final = []
for i in ejer_5.copy():
    if i != 3:
        final.append(i)
print(f"La lista final iterando sobre copia es {final}")

# Y también se pueden ir eliminando elementos (hecho en clase en la correción)
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
for i in ejer_5.copy():
    if i == 3:
        ejer_5.remove(i)
print(f"La lista final eliminando es {ejer_5}")
````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]

diccionario = {"x": [1, 2, 3], "y": [1, 2], "z": 1}
print(f"El diccionario es {diccionario} y es de tipo {type(diccionario)}")
print("-" * 50)

# También se puede hacer con un bucle (visto en clase en la corrección)

ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
dict_11 = {}
for clave, valor in ejer_11:
    if clave not in dict_11:
        dict_11[clave] = []
    dict_11[clave].append(valor)
print(f"El diccionario es {dict_11}")

````

## Ejercicio 12

3. Añade una nueva clave/valor
4. Busca la clave igual a 2 dentro del diccionario
5. Itera la clave y el valor del diccionario con un único for
> ```ejer_12 = {4:78, 2:98, 8:234, 5:29}```

````python
ejer_12 = {4:78, 2:98, 8:234, 5:29}

listaclaves = list(ejer_12.keys())
listaasc = sorted(listaclaves)
print(f"La lista de claves ordenada de manera ascendente es {listaasc}")

listavalores = list(ejer_12.values())
print(f"La lista de valores es {listavalores}")
listavalores.sort(reverse=True)
print(f"La lista descendente de valores es {listavalores}")

print(f"La clave igual a 2 en el diccionario es {ejer_12[2]}")
print("-------------------------------------------------------")
for clave, valor in ejer_12.items():
    print(f"La clave es {clave} y su valor es {valor}")
    
````

## Ejercicio 14
Suma todos los valores del dicionario
> ```ejer_14 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_14 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}
suma = ejer_14[1] + ejer_14[2] + ejer_14[3] + ejer_14[4] + ejer_14[5]
print(f"La suma de los valores del diccionario es {suma}")
print("-" * 50)
# También se puede hacer con un bucle (hecho en la correción de clase)
suma = 0
for i in ejer_14.values():
    suma += i
print(f"La suma es {suma}")
````

## Ejercicio 15
Multiplica todos los valores del diccionario
> ```ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}
mult = ejer_14[1] * ejer_14[2] * ejer_14[3] * ejer_14[4] * ejer_14[5]
print(f"Los valores multiplicados dan un total de {mult}")
print("-" * 50)
# También se puede hacer con un bucle (hecho en la correción de clase)
mult = 1
for i in ejer_14.values():
    mult = mult * i
print(f"La multiplicación es {mult}")
print("-" * 50)

# Se puede importar una librería para este caso (hecho en la correción de clase)

import math

print(f"La multiplicación usando la librería math también da {math.prod(ejer_15.values())}")
````

## Ejercicio 16
1. Crea un set de tres elementos
2. Añade un cuarto
3. Elimina el último elemento añadido

````python
set = {1, 2, 3}
print(f"Mi set es {set}")
set.add(100)
print(f"Ahora el set es {set}")
set.remove(100)
print(f"Después de eliminar, el set es {set}")
set.discard(10) # En este caso, el 10 no se elimina porque no existía, pero no da error
````
