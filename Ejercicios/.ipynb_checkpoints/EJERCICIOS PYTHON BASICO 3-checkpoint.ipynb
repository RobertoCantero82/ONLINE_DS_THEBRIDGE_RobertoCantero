## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C 
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

print(A == B)
print(A != B)
print(A > C)
print(C <= A)
print(B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
input1 = input("Dame un input:")
input2 = input("Dame otro input:")
print(input1 == input2)
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input1 = input("Dame un input:")
input2 = input("Dame otro input:")
input3 = input("Venga, dame un último input:")
print(input1 == input2 and input2 == input3 and input1 == input3)
print(input1 == input2 or input2 == input3 or input1 == input3)
````

## Ejercicio 4
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

1. Solo sería True si todos fuesen True (por el operador and). Como hay un False, el resultado es **False**.
2. El primer paréntesis indica que será True si uno de los valores lo es. En este caso, es True. Con el segundo paréntesis, lo mismo.
   Ya que tenemos dos True, el operador and nos indica que los dos deben ser True para devolver True. También lo es.
   Solo nos queda not True. Con lo que el resultado, que es lo contrario por el not, es **False**.
3. El paréntesis indica que mientras uno de los argumentos sea True, el resultado será True, como así lo es. Ahora, True and False. Como deben ser los dos True para cumplirse y no lo son, el resultado es **False**.
4. Simplificamos el not False con True y el not True con False. Por tanto, True and True and False nos dará como resultado **False**.
