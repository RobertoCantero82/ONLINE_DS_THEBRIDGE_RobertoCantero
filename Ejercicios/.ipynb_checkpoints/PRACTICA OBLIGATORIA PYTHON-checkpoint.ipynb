## PARTE I

En este ejercicio debes simular un chatbot al que le haremos pedidos de pizzas.

1. El chatbot tiene que saludar con un: "Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas  familiares desea?"
2. El ususario tiene que introducir un número de pizzas en una variable llamada 'pizz_familiar'
3. El chatbot debe responder: "Estupendo, ¿cuántas pizzas medianas desea?" y guardar el número de pizzas medianas en otra variable 'pizz_mediana'
4. Además debe responder del chatbot: "Estupendo, se están preparando 'pizz' pizzas familiares y 'pizz_mediana' pizzas medianas. Digame su dirección"
5. El ususario tiene que introducir una direccion en formato String en otra variable llamada 'direcc'
6. El programa debe calcular el monto total
7. Respuesta final del chatbot: "Le mandaremos las '(número total de pizzas)' pizzas a la dirección 'direcc'. Serán <precio total a pagar en euros>. Muchas gracias por su pedido."

```python
pizz_familiar = int(input("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestras pizzas es de 8,95 para la familiar o de 9,90 para la mediana ¿Cuántas pizzas familiares desea?"))
pizz_mediana = int(input("Estupendo, ¿cuántas pizzas medianas desea?"))
direcc = input(f"Estupendo, se están preparando {pizz_familiar} pizzas familiares y {pizz_mediana} pizzas medianas. Dígame su dirección")
pizz = int(pizz_familiar + pizz_mediana)
precio = (pizz_familiar * 8.95) + (pizz_mediana * 9.90)
print(f"Le mandaremos las {pizz} pizzas a la dirección {direcc}. Serán {precio:.2f} euros. Muchas gracias por su pedido")
```

## Parte II

Crea un programa que realice las siguientes acciones:

1. Solicite al usuario que introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3).
2. Convierta esa entrada en una lista de números enteros.
3. Utilizando funciones built-in, determine el número máximo, el número mínimo y la suma total de los números introducidos.
4. Determine la verdad o falsedad de la siguiente afirmación: "El número máximo menos el número mínimo es igual a la suma total dividida por 5".
5. Muestre los resultados utilizando print.

```python
numeros = input("Introduce tres números enteros separados con coma. (Por ejemplo 1, 2, 3)")
print(f"Estos son tus números: {numeros}")
```

```python
print("La lista de los números la puedes consultar debajo de esta línea")
print(numeros.split())
```

```python
print("Debajo puedes ver el número más alto, el más bajo y su suma")
print(max(numeros), min(numeros), sum(numeros))
```

```python
# Da error porque estoy sumando elementos de un string y no enteros. Primero hay que convertirlos y después ya se puede operar con ellos.

# Aquí debajo convierto los strings en enteros. La nueva lista, enteros, será creada a partir de la lista numeros. Para ello, la lista numeros
# se divide por cada ",". Después, con el bucle for, se recorre cada posición de la lista y se transforma en int.

enteros = [int(x) for x in numeros.split(",")]

print("El número más alto es:", max(enteros))
print("El número más bajo es:", min(enteros))
print("La suma de todos los números es:", sum(enteros))
```

```python
afirmacion = (max(enteros) - min(enteros)) == (sum(enteros) / 5)
print("La afirmación es", afirmacion)
```
